```python
import html
import requests
import json
from bs4 import BeautifulSoup
import os
```

```python
url = "https://news.ycombinator.com/newest"
data = requests.get(url)
```

```python
soup = BeautifulSoup(data.text, 'html.parser')
```

```python
soup.prettify()
```

```python
score = soup.find_all('span', class_ = 'score')
submission = soup.find_all('tr', class_ = "athing submission")
```

```python
dict = {}
x = 0
for i in submission:
    temp = i.find('span', class_ = "titleline")
    if score[x]['id'].split('_')[1] == i['id']:
        dict[i['id']] = {'link':temp.a['href'], 'title':temp.a.get_text(strip=True),'score': score[x].text}  

    else:
        dict[i['id']] = {'link':temp.a['href'], 'title':temp.a.get_text(strip=True), 'score': None}
    x += 1
```

```python
with open("data.json", "w") as f:
    json.dump(dict,f,indent=4)
```
